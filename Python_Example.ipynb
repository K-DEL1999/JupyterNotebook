# Very Simple Python Problem from LeetCode

## The Problem

The problem asks you, the programmer, to design an algorithm to count the amount of **active** bits in the binary representation of a given integer. You have to do this for all numbers from 0 to n - n can be an integer from 0 to UNSIGNED_INT_MAX. The values are to be stored in an array and this array is to be returned. It is not a requirement but the problem also challenges you to solve the problem in a linear runtime O(n).  

## The Brute Force Solution

The brute force solution is very simple and intuitive. You just construct the binary representation of the value and count the bits that are _active_. You do this for all values from 0 to n and then store it in a vector. This solution requires O(n) space but also has an O(nlogn) runtime.

### My Implementation

```python
# O(nlogn) method
class Solution(object):
    def countBits(self, n):
        def bits(val): # helper function
            b = 0 # intial count of active bits in binary represenation of value
            while True: 
                if val % 2 == 1 or val / 2 == 0: # checks to see if we increment 
                    b += 1
                val = val // 2 # update value based on integer division
                if val == 0: # if value is 0 we are done
                    break
            return b
        
        output = [0] # initial state of array
        if n == 0: # condition that checks if given n is 0
            return output # return [0]
        else:
            for i in range(1,n+1): # loops from 1 to n 
                output.append(bits(i)) # appends number of bits to array
            return output # returns array
```

```python
S = Solution()
print(S.countBits(11))
```

## The Right Shift Solution

The right shift solution is a dynamic programmin solution. It uses already existing data computed by the algorithm to solve the current problem. The main idea to understand about this algorithm is that when you divide a value by 2, in binary you are performing a right shift. 

#### Example using odd integer

$ 7 / 2 = 3 $
$ 111 >> 1 = 011 $ with remainder 1

#### Example using even integer

$ 6 / 2 = 3 $
$ 110 >> 1 = 011 $ with remainder 0

So from the above example, we see that we can utilize the answer we have for the (current_value)//2 <--- integer division ---- and then add 1 if the current value is odd.

$ ActiveBits(i) = Array[i//2] + (imod2) $

### My Implementation

```python
# O(n) method
class Solution(object): # python class
    def countBits(self, n):  # python function within class
        # O(n)
        Output = [0] # creates a list with 0 as an element
        if n > 0: # conditional statment checking whether n is greater than 0
            for i in range(1,n+1): # a loop that ranges from 1 to n
                Output.append(Output[i//2] + i%2) # appends number of active(1) bits in binary represenation of index
        return Output # returns list
        
```

```python
S = Solution()
print(S.countBits(11))
```
